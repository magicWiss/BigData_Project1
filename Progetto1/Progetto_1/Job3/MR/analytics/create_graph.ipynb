```python
#creazione del grafo indotto dalle relazioni del job3
PATH_ROW_DATA="../output/mr2.txt"
with open(PATH_ROW_DATA,'r') as f:
    lines=f.readlines()


```

```python
#creazione degli insiemi V_user,V_product,E del grafo
from itertools import product
def create_user_nodes(lines):
    user_nodes=[]
    product_nodes=[]
    edges=[]
    limit=200
    conta=0
    for line in lines:
        print(conta,"\t",line)
        if conta==limit:
            break
        
        conta+=1
        row=line.split('\t')
        print(row)
        current_user_nodes=eval(row[0])
        current_prod_nodes=eval(row[1])
        

        
        current_edges=list(product(current_user_nodes,current_prod_nodes))
      
        user_nodes=user_nodes+current_user_nodes
        product_nodes=product_nodes+current_prod_nodes
        edges=edges+current_edges
    user_nodes=set(user_nodes)
    user_nodes= [[x, "User"] for x in user_nodes]
    product_nodes=set(product_nodes)
    product_nodes= [[x, "Prod"] for x in product_nodes]
    edges=set(edges)
    return [user_nodes,product_nodes,edges]




```

```python
graph=create_user_nodes(lines)

user_nodes=graph[0]
prod_nodes=graph[1]
edges=graph[2]

print(edges)

```

```python
import pandas as pd

nodes=user_nodes+prod_nodes
user_nodes=pd.DataFrame(data=user_nodes,columns=["UserId","Class"]).to_csv("UserNodes.csv",index=False)
prod_nodes=pd.DataFrame(data=prod_nodes,columns=["ProductId","Class"]).to_csv("ProdNodes.csv",index=False)
nodes=pd.DataFrame(data=nodes,columns=["ID","Interval"]).to_csv("Nodes.csv",index=False)
edges=pd.DataFrame(data=edges,columns=["Source","Target"]).to_csv("Edges.csv",index=False)
```
