```python
input_path="input_test.text"

with open(input_path,'r') as f:
    lines=f.readlines()

total=len(lines)
print("Totale:",total)
sum=0

for line in lines:
    values=line.split('\t')
    value1=values[1]
    value2=values[2]
    try:
        value1=float(value1)
        value2=float(value2)
    except ValueError:
        pass
    
    if value2==0:
        print('VAL 1:',value1, 'VAL 2:',value2)
        print('Summa corrente:',sum)
    else:
        current_val=value1/value2
        sum=sum+current_val
        print('VAL 1:',value1, 'VAL 2:',value2)
        print('Current value:',current_val)
        print('Somma corrente:',sum)
        
    print("=======")

        

print(sum/total)

```
