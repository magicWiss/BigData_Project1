```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('tempi.csv')



```

```python
X=list(data.columns)[2:]
data_job1=data[data['job']==1]
data_job2=data[data['job']==2]
data_job1=data_job1.sort_values(by="System")
data_job2=data_job2.sort_values(by="System")

```

```python

def plot_data(data,path,title):
    fig, ax = plt.subplots(figsize=(8, 6))

    for _, row in data.iterrows():
        system = row['System']
        values = row[2:]
        ax.plot(X, values, label=system)

    # Set labels and title
    ax.set_xlabel('Dataset-dimension')
    ax.set_ylabel('Execution-time (sec)')
    ax.set_title(title)
    ax.legend()

    plt.savefig(path)
    # Display the plot
    plt.show()
    
    
```

```python
#LOCAL EXECUTION

plot_data(data_job1,'Local_job1.jpeg',"job1-LOCAL")
plot_data(data_job2,'Local_job2.jpeg',"job2-LOCAL")
```

```python
#CLUSTER EXECUTION
data_cluster=pd.read_csv('tempi_cluster.csv')
X=list(data_cluster.columns)[2:]
data_cluster_job1=data_cluster[data_cluster['job']==1]
data_cluster_job2=data_cluster[data_cluster['job']==2]
data_cluster_job1=data_cluster_job1.sort_values(by="System")

data_cluster_job2=data_cluster_job2.sort_values(by="System")

```

```python

plot_data(data_cluster_job1,'Cluster_job1.jpeg',"job1-CLUSTER")
plot_data(data_cluster_job2,'Cluster_job2.jpeg',"job2-CLUSTER")
```
